```python
import pandas as pd
import numpy as np
import os
```

# ラベルの時間とフレーム数を合わせる

```python
text = pd.read_csv('../Annotations/fine_anno_txt/Baun-Augustyn-GrpG-LondonOlympics-2012.txt', delimiter='\t', sep='', header=None, usecols=[2,3,5], names=["start", "end", "label"])
```

```python
# 時間の単位はミリ秒
text.head()
```

```python
action_labels = pd.read_csv('./labels.txt', header=None, sep = '\t')
action_index = {l:i for i,l in enumerate(action_labels[0])}
action_index
```

```python
annos = np.ones(len(text["label"]))
for i in range(len(annos)):
    annos[i] = action_index[text["label"][i]]
labels = pd.Series(annos, name="label_num")
anno_num = pd.concat([text["start"], text["end"], labels], axis=1)
```

```python
start_time = np.array(text["start"])
end_time = np.array(text["end"])
```

```python
images = np.array(range(8290, 46332))
images_list = images*40
images_list.shape
```

```python
labels_frame = np.zeros(len(images_list))
for i, frame in enumerate(images_list):
    idx = np.nonzero((start_time <= frame) * (frame < end_time))[0]
    if len(idx) > 0:
        labels_frame[i] = anno_num["label_num"][idx]
    else:
        labels_frame[i] = "NaN"
```

```python
labels_frame.shape
```

```python
images_list = pd.Series(images_list, name="Frame")
labels_frame = pd.Series(labels_frame, name="Label")
frame_labels = pd.concat([images_list, labels_frame], axis=1, )
```

```python
# NaNを除去する
true_labels = frame_labels.dropna()
true_labels.shape
```
